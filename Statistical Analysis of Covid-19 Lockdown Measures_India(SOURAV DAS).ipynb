```python
# importing pandas as pd 
import pandas as pd
# importing numpy as np 
import numpy as np
# importing matplotlib as plt
import matplotlib.pyplot as plt
# importing seaborn as sns
import seaborn as sns
# importing scipy.stats as stats
import scipy.stats as stats
# importing math library
import math
# importing datetime
from datetime import datetime
```

```python
# loading dataset
Covid = pd.read_csv('India.csv')
Covid
```

```python
# Checking dataframe info
Covid.info()
```

```python
# Using fillna() function for the whole dataframe
Covid = Covid.fillna(0)
Covid
```

```python
# Describing dataframe
Covid.describe()
```

```python
# Changing dataframe's date format
Covid['date'] = pd.to_datetime(Covid['date'], format = '%d-%m-%Y')

# Checking dataframe
Covid.head(10)
```

```python
# Drawing lineplot
plt.plot(Covid['date'], Covid['confirmed'], linewidth=2)
plt.title('Confirmed Cases Datewise')
plt.xlabel('Date')
plt.ylabel('Confirmed Cases')
plt.show()
```
